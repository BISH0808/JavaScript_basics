# React Хуки


* Хуки - это некоторые функции, которые предоставляет React

* Эти функции всегда начинаются со слова **Use**

* Хуки можно использовать либо в функциональных компонентах, либо в своих собственных хуках. То есть на основании стандартных онлайн хуков, которые представлены ниже, можно делать свои собственные хуки с различным функционалом.

image.png

* Хуки можно использовать только на верхнем уровне вложенности, т.е. нельзя вкладывать хуки внутри какой-то функции, нельзя вкладывать хуки в условия, циклы, только на верхнем уровне компонента или *кастомного* (своего) хука.
