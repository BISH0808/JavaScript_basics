# Как работают классические клиентские приложения (Multi page Application)

* Создается несколько html файлов
* Каждый html файл обладает каким-то своим содержимым, какими-то скриптами (js-скриптами), каждый html файл - как отдельная страница в нашем приложении
* между страницами мы переключаемся с помощью специальных ссылок


image.png

* Такой подход называется Multi Page Application.

* Такой подход имеет как и свои плюсы, так и свои минусы.

* Из минусов:
  * каждый раз при переходе на новую страницу мы загружаем полностью новый html файл, при этом некоторые части, например, навигационная панель, левое или правое меню, footer, они остаются прежними, но мы каждый раз подкачиваем их заново
  * без грамотной оптимизации пользователь будет видеть (особенно при медленном интернете) переходы между страницами, обновление окна бразера
  

# Как работает Single Page Application? (приложение создаваемое при помощи React)

* во всем приложении имеем 1 html файл. Как правило, это index.html
* это минималистичный html файл, которые имеет тэги #body и соответственно #head, ну, и какие-либо метатеги
* все содержимое приходит из JS, т.е. как таковой html разметки мы не пишем
* содержимым страницы можем как хотим
* быстрая работа (все JS файлы подгрузили изначально), footer и другое мы подгрузили изначально
* первичную отрисовку можно оптимизировать, для этого нужно настроить WebPack, использовать LazyLoading и различные другие хитрости

# Как работает React?

* React работает по принципу компонентного подхода: мы создаем какое-то кол-во компонентов и используем их как кирпичики. Вкладывая один компонент в другой мы создаем достаточно сложный интерфейс.
* Компоненты можно переиспользовать.
* При таком подходе есть корневой компонент, который монтируется в index.html файл. Чаще всего такой компонент в контексте React называется App.




* Как вы можете видеть на рисунке выше компонет *App* содержит форму для создания поста и список постов. В свою очередь список постов состоит из n-го числа отдельно взятых постов, а форма для создания содержит в себе Input и кнопку.

# Чем еще React упрощает жизнь?

Разработка приложения без фреймворка заставляет использовать DOM - дерево. Среди операций по DOM-дереву: получить элемент по классу, по ID и другое. То есть, разработчику приходилось больше думать о такой рутине и меньше о логике работы приложения. React позволяет разработчику больше сосредотачиваться на логике по работе с данными, при этом самостоятельное взаимодействие с DOM-деревом минимизировано. Разработчик работает с данными, говорит React, где, что должно находиться, в свою очередь React сам следит за обновлением интерфейса. Т.е. вручную изменять DOM-дерево нужно, т.е. выизменяете данные, а React сам следит за интерфейсом. За React стоят сложные оптимизированные механизмы, такие как согласование, React Fiber, render функция.

image.png

React под копотом строит виртуальное дом дерево, это более легковесная копия обычного DOM дерева в браузере.

image.png

Сами рзработчики предпочиатают название DOM-дереву название "дерево элементов", посколько оно строится не только в дереве, а DOM означает, что дерево строится в браузере.

image.png

* Когда в каких-то узлах дерева произошли изменения, они не переносятся в дом дерево сразу, вместо этого React строит новое дерево элементов с обновленными значениями и сравнивает это дерево с предыдущими. Это достаточно сложные алгоритмы и сложные механизмы, и эта стадия называется согласоваине.
* За согласование отвечает ядро React, оно выполняется независимо от среды выполнения, будь то браузер или телефон.


image.png

* После того как React нашел разницу между этими 2мя деревьями просиходит фазы визуализации, рендеринга, отрисовки, т.е. мы видим обновленные данные на странице. При этом для каждого изменения React устанавливает свою приоритетность: более приоритетные изменения он вносит быстрее, менее приоритетные медленнее. За счет чего мы видим плавную картинкус максимальной высотой частоты кадров.
