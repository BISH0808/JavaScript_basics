# Как работают классические клиентские приложения (Multi page Application)

* Создается несколько html файлов
* Каждый html файл обладает каким-то своим содержимым, какими-то скриптами (js-скриптами), каждый html файл - как отдельная страница в нашем приложении
* между страницами мы переключаемся с помощью специальных ссылок


image.png

* Такой подход называется Multi Page Application.

* Такой подход имеет как и свои плюсы, так и свои минусы.

* Из минусов:
  * каждый раз при переходе на новую страницу мы загружаем полностью новый html файл, при этом некоторые части, например, навигационная панель, левое или правое меню, footer, они остаются прежними, но мы каждый раз подкачиваем их заново
  * без грамотной оптимизации пользователь будет видеть (особенно при медленном интернете) переходы между страницами, обновление окна бразера
