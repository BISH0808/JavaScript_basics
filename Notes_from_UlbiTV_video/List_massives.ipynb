* Когда создаете списки, обязательным условием является указание некоторого ключа, при этом этот ключ должен хранить какое-то уникальное значение. Как правило, это id объекта, который мы используем.

* Ключ должен быть всегда статичным и уникальным. Ключи позволяют алгоритмам React наиболее эффективно делать Rendering и отрисовывать не весь список, а только те элементы, в которых произошли изменения.
